```python
import ast
import pandas as pd

def has_labels(x):
    if pd.isna(x):
        return False
    if isinstance(x, str):
        x = x.strip()
        if x == "" or x == "{}":
            return False
        try:
            d = ast.literal_eval(x)
        except Exception:
            return False
    elif isinstance(x, dict):
        d = x
    else:
        return False
    return len(d) > 0

df = df[df["weighted_skin_condition_label"].apply(has_labels)].copy()
```

## Create a clean label column

- Take the top three conditions and make a new column with a list
- If there is less than three conditions, just include the one/two conditions

```python
import ast
def topk_labels(x, k=3, min_w=None):
    # x can be a dict or a stringified dict
    if pd.isna(x):
        return []
    if isinstance(x, str):
        x = x.strip()
        if x == "" or x == "{}":
            return []
        try:
            d = ast.literal_eval(x)
        except Exception:
            return []
    elif isinstance(x, dict):
        d = x
    else:
        return []

    if not d:
        return []

    items = sorted(d.items(), key=lambda t: t[1], reverse=True)
    if min_w is not None:
        items = [t for t in items if float(t[1]) >= min_w]
    return [name for name, w in items[:k]]

df["labels_top3"] = df["weighted_skin_condition_label"].apply(lambda s: topk_labels(s, k=3, min_w=None))
```

## Convert rare labels (less than 20 examples in the dataset) into an Other label

- This will help the model deal with class imbalance
- Otherwise there will rare conditions with not enough data to train on properly

```python
from collections import Counter

# Count all labels
all_labels = [lab for labs in df["labels_top3"] for lab in labs]
cnt = Counter(all_labels)

MIN_COUNT = 20
keep = {lab for lab, c in cnt.items() if c >= MIN_COUNT}

def replace_with_other(labs):
    new_labs = []
    has_other = False

    for l in labs:
        if l in keep:
            new_labs.append(l)
        else:
            has_other = True

    # Add "Other" only once if any rare label existed
    if has_other:
        new_labs.append("Other")

    return new_labs

df["labels_top3_kept"] = df["labels_top3"].apply(replace_with_other)
```

```python
# Build label vocabulary from keep + explicitly add Other
labels_sorted = sorted(list(keep)) + ["Other"]
label2idx = {l: i for i, l in enumerate(labels_sorted)}

C = len(labels_sorted)

def to_multihot(labs, C):
    y = [0] * C
    for l in labs:
        # skip anything unexpected rather than crashing
        if l in label2idx:
            y[label2idx[l]] = 1
    return y

Y = df["labels_top3_kept"].apply(lambda labs: to_multihot(labs, C))
```

```python
import os
import numpy as np
import pandas as pd
from PIL import Image

import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms

# -------------------------
# CONFIG
# -------------------------
IMAGE_ROOT = "data/raw"
CASES_CSV = "data/raw/scin_cases.csv"
CASE_COL = "case_id"
SEED = 42
BATCH_SIZE = 32
NUM_WORKERS = 2
PIN_MEMORY = torch.cuda.is_available()

# -------------------------
# 1) Index all PNG files (recursive)
# -------------------------
all_png = []
for root, _, files in os.walk(IMAGE_ROOT):
    for f in files:
        if f.lower().endswith(".png"):
            all_png.append(os.path.join(root, f))

print("PNG files found:", len(all_png))
png_stems = [os.path.splitext(os.path.basename(p))[0] for p in all_png]
stem2path = {stem: p for stem, p in zip(png_stems, all_png)}
stem_set = set(png_stems)

print("Example png stems:", png_stems[:5])

# -------------------------
# 2) Load cases and extract stems from image path columns
# -------------------------
cases = pd.read_csv(CASES_CSV)

img_path_cols = [c for c in ["image_1_path", "image_2_path", "image_3_path"] if c in cases.columns]
if not img_path_cols:
    raise ValueError("Expected image_1_path/image_2_path/image_3_path in scin_cases.csv but didn't find them.")

def path_to_stem(x):
    if pd.isna(x):
        return None
    s = str(x).strip()
    base = os.path.basename(s)
    stem, _ = os.path.splitext(base)
    return stem if stem != "" else None

for c in img_path_cols:
    cases[c + "_stem"] = cases[c].apply(path_to_stem)

# Count matches per stem column
for c in img_path_cols:
    m = cases[c + "_stem"].isin(stem_set).sum()
    print(f"Matches via {c}_stem:", int(m))

# -------------------------
# 3) Expand cases to one-row-per-image (up to 3 images per case)
# -------------------------
rows = []
for c in img_path_cols:
    tmp = cases[[ "case_id", c + "_stem" ]].rename(columns={c + "_stem": "image_stem"}).copy()
    tmp["image_slot"] = c  # which of the 3 images
    rows.append(tmp)

cases_img = pd.concat(rows, ignore_index=True)
cases_img = cases_img.dropna(subset=["image_stem"]).copy()

# keep only stems that exist on disk
cases_img = cases_img[cases_img["image_stem"].isin(stem_set)].copy()
cases_img["filepath_full"] = cases_img["image_stem"].map(stem2path)

print("Total case-image rows matched to disk:", len(cases_img))
print(cases_img.head())

# -------------------------
# 4) Attach Y to df and merge with cases_img on case_id
# -------------------------
if not isinstance(df, pd.DataFrame):
    df = pd.DataFrame(df)
if CASE_COL not in df.columns:
    raise ValueError(f"df must contain '{CASE_COL}'. Current columns: {list(df.columns)}")

df = df.copy()

Y_list = Y.tolist() if isinstance(Y, pd.Series) else list(Y)
if len(Y_list) != len(df):
    raise ValueError(f"Length mismatch: len(df)={len(df)} vs len(Y)={len(Y_list)}")

df["y_vec"] = [np.array(v, dtype=np.float32) for v in Y_list]

merged = df.merge(cases_img[["case_id", "filepath_full"]], left_on=CASE_COL, right_on="case_id", how="inner")

# If a case has multiple images, this duplicates y_vec across images (usually what you want)
merged = merged[merged["y_vec"].apply(lambda v: v.sum() > 0)].copy()

print("Merged usable rows (images):", len(merged))
print(merged[[CASE_COL, "filepath_full"]].head())

# -------------------------
# 5) Train/Val/Test split by CASE (prevents leakage across multiple images of same case)
# -------------------------
unique_cases = merged[CASE_COL].drop_duplicates().to_numpy()
rng = np.random.default_rng(SEED)
rng.shuffle(unique_cases)

n_cases = len(unique_cases)
n_test = int(0.15 * n_cases)
n_val  = int(0.15 * n_cases)

test_cases = set(unique_cases[:n_test])
val_cases  = set(unique_cases[n_test:n_test+n_val])
train_cases= set(unique_cases[n_test+n_val:])

df_train = merged[merged[CASE_COL].isin(train_cases)].reset_index(drop=True)
df_val   = merged[merged[CASE_COL].isin(val_cases)].reset_index(drop=True)
df_test  = merged[merged[CASE_COL].isin(test_cases)].reset_index(drop=True)

print("Split sizes (images):", len(df_train), len(df_val), len(df_test))

# -------------------------
# 6) Transforms
# -------------------------
train_tfms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
])

eval_tfms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

# -------------------------
# 7) Dataset + DataLoaders
# -------------------------
class SCINDataset(Dataset):
    def __init__(self, df, transform=None):
        self.df = df
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        img = Image.open(row["filepath_full"]).convert("RGB")
        if self.transform:
            img = self.transform(img)
        y = torch.tensor(row["y_vec"], dtype=torch.float32)
        return img, y

train_ds = SCINDataset(df_train, transform=train_tfms)
val_ds   = SCINDataset(df_val, transform=eval_tfms)
test_ds  = SCINDataset(df_test, transform=eval_tfms)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True,
                          num_workers=NUM_WORKERS, pin_memory=PIN_MEMORY)
val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=NUM_WORKERS, pin_memory=PIN_MEMORY)
test_loader  = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=NUM_WORKERS, pin_memory=PIN_MEMORY)

# -------------------------
# 8) Sanity check
# -------------------------
xb, yb = next(iter(train_loader))
print("X batch:", xb.shape)
print("Y batch:", yb.shape)
print("Label sums (should be 1-3):", yb.sum(dim=1)[:10])
print("Num classes:", int(yb.shape[1]))
```

```python
import torch
import torch.nn as nn
from torchvision import models

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
C = 59  # from your output

model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
model.fc = nn.Linear(model.fc.in_features, C)
model = model.to(device)

criterion = nn.BCEWithLogitsLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=3e-4, weight_decay=1e-4)
```

```python
def precision_recall_at_k(probs, targets, k=3):
    # probs: (B, C) sigmoid outputs
    # targets: (B, C) multi-hot 0/1
    topk = torch.topk(probs, k=k, dim=1).indices  # (B, k)

    pred = torch.zeros_like(targets)
    pred.scatter_(1, topk, 1.0)

    tp = (pred * targets).sum(dim=1)
    prec = tp / k
    denom = targets.sum(dim=1).clamp(min=1.0)
    rec = tp / denom

    return prec.mean().item(), rec.mean().item()
```

```python
from tqdm import tqdm

def run_epoch(model, loader, train=True):
    model.train() if train else model.eval()
    total_loss = 0.0
    total_prec3 = 0.0
    total_rec3 = 0.0
    n_batches = 0

    for xb, yb in tqdm(loader, leave=False):
        xb = xb.to(device, non_blocking=True)
        yb = yb.to(device, non_blocking=True)

        with torch.set_grad_enabled(train):
            logits = model(xb)
            loss = criterion(logits, yb)

            if train:
                optimizer.zero_grad(set_to_none=True)
                loss.backward()
                optimizer.step()

        probs = torch.sigmoid(logits.detach())
        p3, r3 = precision_recall_at_k(probs, yb, k=3)

        total_loss += loss.item()
        total_prec3 += p3
        total_rec3 += r3
        n_batches += 1

    return (
        total_loss / n_batches,
        total_prec3 / n_batches,
        total_rec3 / n_batches
    )

EPOCHS = 5

for epoch in range(1, EPOCHS + 1):
    train_loss, train_p3, train_r3 = run_epoch(model, train_loader, train=True)
    val_loss, val_p3, val_r3 = run_epoch(model, val_loader, train=False)

    print(
        f"Epoch {epoch:02d} | "
        f"train loss {train_loss:.4f} P@3 {train_p3:.3f} R@3 {train_r3:.3f} | "
        f"val loss {val_loss:.4f} P@3 {val_p3:.3f} R@3 {val_r3:.3f}"
    )
```

```python
test_loss, test_p3, test_r3 = run_epoch(model, test_loader, train=False)
print(f"TEST | loss {test_loss:.4f} P@3 {test_p3:.3f} R@3 {test_r3:.3f}")
```
